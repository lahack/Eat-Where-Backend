```python
import math
def Dist(row, input_labels): # to be modified
    distance = 0
    for j in range(len(row)):
        if j in range(8, 17): # 9 options
            if (row[j] == 1 and input_labels[j] == 0):
                distance += 100
            elif row[j] == 1 and input_labels[j] > 0:
                distance -= 30 * input_labels[j]
            elif row[j] == 1 and input_labels[j] < 0:
                distance += 100 + (-1) * 500 * input_labels[j]
            else:
                continue
        elif j in range(1, 6): # 5 options
            if (row[j] == 0 and input_labels[j] == 1):
                distance += 1000
            else:
                continue
        else:
            dist = (row[j] - input_labels[j])*(row[j] - input_labels[j])
            distance += dist
    distance = math.sqrt(distance)
    return distance
# fake_user_input = [1, 2, 3, 4, 5, 6, 7, 8]
# row = [1, 2, 3, 4, 5, 6, 7, 8]
# print(Dist(row,fake_user_input))
```

```python
def Get_Input_Labels(user_input):
    input_labels = []
    avg_price = (user_input[0][0] + user_input[0][1]) / 2
    input_labels.append(avg_price) # price
    if (user_input[1] == 'SINGLE'):
        input_labels.extend([1, 0, 0, 0, 0])
    elif user_input[1] == 'DATE':
        input_labels.extend([0, 1, 0, 0, 0])
    elif user_input[1] == 'BUSINESS':
        input_labels.extend([0, 1, 0, 0, 0])
    elif user_input[1] == 'CASUAL':
        input_labels.extend([0, 1, 0, 0, 0])
    else:
        input_labels.extend([0, 0, 1, 0, 0]) # function
    input_labels.extend([user_input[2][0], user_input[2][1]]) # location
    # type
    types = user_input[3]
    if ('JAPANESE' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('KOREAN' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('CHINESE' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('THAI' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('MEXICAN' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('ITALIAN' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('FAST_FOOD' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('MEDI' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
    if ('INDI' in types):
        input_labels.append(1)
    else:
        input_labels.append(0)
   
    # spicy
    input_labels.append(user_input[4])
    input_labels.append(5)
    return input_labels
```
